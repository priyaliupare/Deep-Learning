```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('multisat.csv')
```

```python
dataset
```

```python
x1 = dataset[['SAT','Rand 1,2,3']]
```

```python
x1
```

```python
y = dataset['GPA']
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(x1)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
test = pd.DataFrame([[1900,2],[1600,1]],columns=['SAT','Rand 1,2,3'])
```

```python
test
```

```python
test = sm.add_constant(test)
```

```python
test
```

```python
model.predict(test)
```
